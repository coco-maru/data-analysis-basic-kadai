```python
#matplotlibライブラリをインポートする
import matplotlib.pyplot as plt
#numpyライブラリをインポートする
import numpy as np

#グラフを描画する領域を指定する
fig,axes = plt.subplots(2,3,tight_layout=True)

#折れ線グラフを作成する
a = [1,4,6,7,-2]
axes[0,0].plot(a)

#sin関数グラフを作成する
x = np.linspace(0,20,150)
y = np.sin(x)
axes[0,1].plot(x,y)
plt.xlim(10,40)

#ヒストグラムを作成する
b = np.random.normal(10,5,1000)
axes[0,2].hist(b,bins=20)
axes[0,2].set_xlim(-10,30)

#散布図を描画する
np.random.seed(30)
d = np.random.rand(30)
e = np.random.rand(30)
colors =np.random.rand(30)
area=np.random.rand(30)*100
axes[1,0].scatter(d,e,s=area,c=colors,alpha=0.5)

#グラフを表示する
plt.show()
```
